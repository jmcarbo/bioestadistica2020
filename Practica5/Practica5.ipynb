```r
muestreo <- function (a) {
experiment<-rnorm(a)
hist (experiment,prob=TRUE)
x=seq(-4,4,length=200)
th=dnorm(x)
lines (x,th)
}

normal <-function (a,b) {
x=seq(a-4*b,a+4*b, length=200)
y=dnorm(x,a,b)
plot(x,y,type="l",lwd=2, col="red")
}

tstudent <-function (df) {
x=seq(-4,4, length=200)
y=dt(x,df)
dflab <- paste("df =",df)
plot(x,y,type="l",lwd=2, col="red", ylim=c(0.0,0.40),
     main=dflab)
}

qnormGRAPH <-function (a){
x=seq(-4,4,length=200)
y=dnorm(x)
plot(x,y,type="l",lwd=2, col="red")

x=seq(-4,qnorm(a),length=200)
y=dnorm(x)
polygon(c(-4,x,qnorm(a)),c(0,y,0),col="gray")

text (0,0.1,format (qnorm(a),digits=4))
}

dnormGRAPH <-function (a){
x=seq(-4,4,length=200)
y=dnorm(x)
plot(x,y,type="l",lwd=2, col="red")

x=seq(-4,a,length=200)
y=dnorm(x)
polygon(c(-4,x,a),c(0,y,0),col="gray")

text (0,0.1,format(pnorm(a),digits=4))
}

pnormGRAPH <- function(a) {
x<-seq(-4,4,length=200)
y=pnorm(x)
plot(x,y,type="l",lwd=2)

lines(c(-4,a),c(pnorm(a),pnorm(a)),col="red")
lines(c(a,a),c(0,pnorm(a)),col="red")

text (a,0,format(a,digits=4), pos=4)
text (-4,pnorm(a)+0.02,format(pnorm(a), digits=4), pos=4)
}
```

La distribución normal puede representarse gráficamente usando el comando “normal(x,sd)”, donde x es la media de la variable aleatoria y sd su desviación estándar. Prueba a llamarla con distintos valores de x y sd.  > normal (0,1) > normal (10,5) > normal (3082,450) Como verás, la forma no cambia, solo la escala.

La distribución normal permite calcular la probabilidad de que la variable tome valores dentro de un cierto intervalo. Esto puede hacerse calculando el área encerrada bajo la curva, en el gráfico de densidad de probabilidad o leyendo la altura en un gráfico de probabilidad acumulada. Por ejemplo, para una variable con media 560 y desviación estándar 24, calcula la probabilidad de obtener valores menores de 590.  # calculamos el valor estandarizado restando la media y dividiendo por sd > (590-560)/24Se obtiene como valor normalizado 1.25 #calculamos la probabilidad acumulada como  > pnorm(1.25)

Este valor nos indica la probabilidad de obtener valores menores y corresponde a la superficie encerrada bajo la curva desde menos infinito hasta 1.25 en el gráfico de densidad. Para verlo gráficamente usando los comandos # como área bajo la curva, desde menos infinito hasta este valor > dnormGRAPH(1.25) #como valor de la integral de la curva anterior > pnormGRAPH(1.25)
